## Cleaning Data for Life_Expectancy and Fertility Datasets


### In a first review, both datasets contains technically the same issues, so this notebook words well with the both datasets, just replace the const variables for reading the database and saving it.

```python
import pandas as pd
import numpy as np
```

### Configuring the const variables

```python
#file = ["life_expectancy.csv","Clean_Life_Expectancy2.csv","Clean_Life_Expectancy_Melt2.csv",['Country_Name','Country_Code', 'Year','Life_Expectancy']]
file = ["fertility.csv","Clean_fertility.csv","Clean_fertility_Melt.csv",['Country_Name','Country_Code', 'Year','Fertility']]
```

## Starting the cleaning process

### Verifiying if the dataset has null data 

```python
life_df = pd.read_csv(file[0])
life_df.head()
```

### Lookup to the data

```python
life_df.info()
```

```python
life_df.describe()
```

### Deleting null columns (2013 year)

```python
del life_df['2013']
```

### Deleting null rows and ensure that the data is numeric

```python
numeric=[]
for x in range(0,53):
    numeric.append(str(x + 1960))
 
life_df[numeric] = life_df[numeric].apply(pd.to_numeric, errors='coerce')
life_df = life_df.dropna()
print(life_df.shape)
life_df.head()
```

### Deleting duplicates rows

```python
life_df = life_df.drop_duplicates(subset=['Country_Code'])
```

### Deleting no countries data (Regions like Central_Europe_and_the_Baltics and East_Asia_&_Pacific_(developing_only))

```python
countries = life_df["Country_Name"]
#No one country has "()",":" in its name but is possible having _ since costa_rica is a country, just was wrote _ instead of space
pattern = '^[A-Za-z\._\s]*$'

mask = countries.str.contains(pattern)

mask_inverse = ~mask

invalid_countries = countries.loc[mask_inverse]
life_df = life_df[mask]

print(life_df.shape)
print(invalid_countries)
```

### Creating a new dataset replacing the years columns as a rows 

```python
life_df_melt = pd.melt(life_df,id_vars=["Country_Name","Country_Code"])
life_df_melt.columns = file[3]
life_df_melt.head()
```

### Saving both datasets 

```python
life_df.to_csv(file[1],sep=",")
life_df_melt.to_csv(file[2],sep=",")
```
